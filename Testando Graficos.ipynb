```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

tabela = pd.read_csv("train.csv")
soma = tabela["SalePrice"].sum()
media = soma / len(tabela["SalePrice"])
quantidade = tabela["MSZoning"].value_counts()

# Criando o gráfico de barras
plt.figure(figsize=(6, 4))  # Define o tamanho da figura
quantidade.plot(kind='kde')  # Plota o gráfico de barras

plt.title('Contagem de Valores Únicos em "MSZoning"')  # Define o título
plt.xlabel('Tipo de Zoneamento')  # Define o rótulo do eixo x
plt.ylabel('Contagem')  # Define o rótulo do eixo y
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade

plt.show()  # Exibe o gráfico

#display (tabela)
#display(soma)
#display(media)
#display(quantidade)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

tabela        = pd.read_csv("train.csv")
soma          = tabela["SalePrice"].sum()
media         = soma / len(tabela["SalePrice"])
quantidade    = tabela["MSZoning"].value_counts()
porcentagem   = quantidade.apply(lambda x: (x / len(tabela["SalePrice"])) * 100)
porcentagem.plot.bar()
plt.show()
display (tabela)
display (soma)
display (f"Média em porcentagem: {media:.2f}%")
display (quantidade)
display (porcentagem)
```
