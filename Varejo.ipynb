```python
import pandas as pd
# Arrumando as colunas de Data e invertendo os valores tipo Mês para Dia
varejo = pd.read_csv('VAREJO.csv')
varejo['Date'] = pd.to_datetime(
      varejo['Date'],
        format='%m/%d/%Y').dt.strftime('%d/%m/%Y')
display(varejo)
# Salva as alterações em um novo arquivo
varejo.to_csv('VAREJO_formatado.csv', index=False)
```

```python
# Transfromando em Valores em Inteiros
import pandas as pd
import numpy as np
varejo = pd.read_csv('VAREJO.csv')
varejo[['Customer_ID','Transaction_ID','Zipcode','Ratings','Age','Total_Amount',
        'Amount','Total_Purchases','Year']] = varejo[
          ['Customer_ID',
            'Transaction_ID',
              'Zipcode','Ratings', 'Age','Total_Amount',
        'Amount','Total_Purchases','Year']].fillna(0).astype(int)
varejo['Phone'] = varejo[
        'Phone'].fillna(0).astype(int)
varejo['Date'] = pd.to_datetime(
      varejo['Date'],
        format='%m/%d/%Y').dt.strftime('%d/%m/%Y')
display(varejo)
varejo.to_csv('VAREJO_formatado.csv', index=False)
```

```python
# Importando Bibliotecas
import pandas as pd
import numpy as np

varejo = pd.read_csv('VAREJO.csv')

# Converte colunas para inteiros e preenche NaN com 0
cols_int = ['Customer_ID',
              'Transaction_ID',
                'Zipcode',
                  'Ratings',
                    'Age',
                      'Total_Purchases',
                        'Year',
                          'Phone']
varejo[cols_int] = varejo[cols_int].fillna(0).astype(int)

# Formata 'Total_Amount' e 'Amount' com 3 casas decimais
varejo[['Total_Amount',
          'Amount']] = varejo[[
                        'Total_Amount',
                          'Amount']].fillna(0).astype(float).apply(
                                    lambda x: x.apply(lambda y: f'{y:.3f}'))
# Formata a data
varejo['Date'] = pd.to_datetime(varejo['Date'],
                                  format='%m/%d/%Y').dt.strftime('%d/%m/%Y')

display(varejo)
varejo.to_csv('VAREJO_formatado.csv', index=False)
```
